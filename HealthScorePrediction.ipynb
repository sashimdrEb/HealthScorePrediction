```python
import pandas as pd

df = pd.read_csv("/Users/ebpearls/.cache/kagglehub/datasets/simaanjali/diabetes-classification-dataset/versions/1/Diabetes Classification.csv")
df.tail(5)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Step 1: Create a synthetic HealthScore (example formula)
df['HealthScore'] = 100 - (0.5 * df['BMI']) - (0.3 * df['LDL'])

# Step 2: Preprocess
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
X = df.drop('HealthScore', axis=1)
y = df['HealthScore']
print(X.shape)
```

```python
# import numpy as np

# # Step 3: Train model
# model = RandomForestRegressor()
# model.fit(X, y)

# # Predict for a new patient
# new_patient = [[45, 1, 24.5, 190, 150, 40, 120, 0.9, 15]]  # No Diagnosis needed if not used
# print(np.array(new_patient).shape)  # Should be (1, 11)

# print("Predicted Score:", model.predict([new_patient])[0])


import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import time

# Assuming X and y are already defined
print("Starting training...")
print(f"Input shape: {X.shape}")  # Should be (n_samples, 11)

# Step 3: Train model with verbose output
model = RandomForestRegressor(
    n_estimators=100,  # Number of trees
    verbose=2,        # Shows tree-building progress (0=silent, 1=some, 2=detailed)
    n_jobs=-1         # Use all CPU cores
)

start_time = time.time()
print("\nTraining Random Forest model...")
model.fit(X, y)
training_time = time.time() - start_time

print(f"\nTraining completed in {training_time:.2f} seconds")
print(f"Number of trees trained: {len(model.estimators_)}")

# Feature importance (optional logging)
print("\nFeature importances:")
for i, (col, importance) in enumerate(zip(X.columns, model.feature_importances_)):
    print(f"{i+1}. {col}: {importance:.4f}")

# Predict for a new patient
new_patient = [[45, 1, 24.5, 190, 150, 40, 120, 0.9, 15, 0, 0]]  # Ensure 11 features
print("\nPrediction input shape:", np.array(new_patient).shape)  # Should be (1, 11)

prediction = model.predict(new_patient)
print("\nPredicted Score:", prediction[0])
```
